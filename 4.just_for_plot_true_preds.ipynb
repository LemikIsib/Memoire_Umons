```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
attack = ("Normal", "Generic", "Exploits", "Fuzzers", "DoS",
          "Reconnaissance", "Analysis", "Backdoor", "Shellcode", "Worms")
counts = {
    'True': (36844, 18871, 11132, 6062, 4089, 3496, 677, 583, 378, 44),
    'Pred': (19204, 18268, 12962, 22208, 3342, 895, 322, 3102, 1852, 21),
}

x = np.arange(len(attack))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(12,5))

for attribute, measurement in counts.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3, rotation=90)
    multiplier += 1

ax.set_ylabel('count')
ax.set_xlabel('attack_cat')
ax.set_xticks(x + width, attack)
ax.legend(loc='upper right', ncols=3)
plt.savefig("xgb_preds2.pdf", dpi=500, facecolor="w", bbox_inches="tight")

plt.show()
```
